### 1. Обобщим собственную предобработку данных в отдельный класс для удобства

```python
class DataPreprocessor(BaseEstimator, TransformerMixin):
  def _prepare_total_spent(self, X):
    """
    Заменяем пустые строки на '0.0' в колонке TotalSpent и преобразуем к типу float64
    :param X: pd.DataFrame
    :return: X
    """
    X['TotalSpent'] = X['TotalSpent'].replace(' ', '0.0').astype('float64')
    return X

  def fit(self, X, y=None):
    """
    Подготовка класса для трансформаций, тут ничего не делаем
    :param X: pd.DataFrame
    :return: self
    """
    return self

  def transform(self, X):
    """
    Выполняем базовую предобработку данных
    :param X: pd.DataFrame
    :return: X
    """
    X = X.copy()
    X = self._prepare_total_spent(X)
    return X
```

```python
FEATURE_CONFIG = {
    # Числовые признаки, которые будем масштабировать
    'numeric': [
        'ClientPeriod',
        'MonthlySpending',
        'TotalSpent'
    ],

    # Бинарные признаки, которые уже имеют тип int64
    'binary': [
        'IsSeniorCitizen'
    ],

    # Категориальные признаки, которые будем кодировать
    'categorical': [
        'Sex',
        'HasPartner',
        'HasChild',
        'HasMultiplePhoneNumbers',
        'HasInternetService',
        'HasOnlineSecurityService',
        'HasOnlineBackup',
        'HasDeviceProtection',
        'HasTechSupportAccess',
        'HasOnlineTV',
        'HasMovieSubscription',
        'HasContractPhone',
        'IsBillingPaperless',
        'PaymentMethod'
    ],

    # Целевая переменная
    'target': [
        'Churn'
    ]
}
```

### 3. Настраиваем пайплайн

```python
def load_data(feature_config):
  # Все признаки, используемые в обучении
  feature_cols = (feature_config['numeric'] +
                  feature_config['binary'] +
                  feature_config['categorical'])

  # Целевая переменная
  target_col = feature_config['target']

  # Загружаем данные
  df_train = pd.read_csv('./train.csv')
  df_test = pd.read_csv('./test.csv')

  # Возвращаем в удобном для использования формате
  return df_train[feature_cols], df_train[target_col], df_test[feature_cols]
```

```python
col_transformer = ColumnTransformer(
    transformers=[
        ('scaler', StandardScaler(), FEATURE_CONFIG['numeric']),
        ('encoder', OneHotEncoder(drop='first', sparse_output=False), FEATURE_CONFIG['categorical']),
        ('binary', 'passthrough', FEATURE_CONFIG['binary'])
    ]
)

logreg_pipeline = Pipeline(
    steps=[
    ('clean', DataPreprocessor()),
    ('preproc', col_transformer),
    ('model', LogisticRegression(max_iter=1000, random_state=42))
])

param_grid = {
    'model__C' : [100, 10, 1, 0.1, 0.01, 0.001],
    'model__solver' : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
}
```

### 4. Поиск по сетке с помощью `GridSearchCV`

```python
X_train, y_train, X_test = load_data(FEATURE_CONFIG)
```

```python
grid_search = GridSearchCV(
    logreg_pipeline,
    param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)
```

```python
grid_search.fit(X_train, y_train)
```

### 5. Анализ результатов поиска по сетке

#### 5.1 Построим тепловую карту

```python
regularization = grid_search.cv_results_['param_model__C'].data
solvers = grid_search.cv_results_['param_model__solver'].data
auc_score = grid_search.cv_results_['mean_test_score']
```

```python
sns.set_style('white')
plt.figure(figsize=(7, 5))

heatmap_data = pd.DataFrame(data={'C': regularization, 'Solver': solvers, 'AUC': auc_score})
heatmap_data = heatmap_data.pivot(index='C', columns='Solver', values='AUC')

sns.heatmap(heatmap_data, annot=True, cmap=cmap).set_title('AUC for LogRegression')
```

#### 5.2 Выведем оптимальные гиперпараметры логистической регрессии

```python
best_C = grid_search.best_params_['model__C']
best_solver = grid_search.best_params_['model__solver']
best_score = grid_search.best_score_

print('Optimal Inverse of regularization strength ', best_C)
print('Optimal Algorithm                          ', best_solver)
print('AUC-ROC                                    ', best_score)
```

### 6. Создание итоговой модели

```python
classifier = Pipeline(
    steps=[
    ('clean', DataPreprocessor()),
    ('preproc', col_transformer),
    ('model', LogisticRegression(max_iter=1000, random_state=42, C=best_C, solver=best_solver))
])
```

```python
X_train, y_train, X_test = load_data(FEATURE_CONFIG)
```

```python
classifier.fit(X_train, y_train)
```

#### 6.1 Предсказания логистической регрессии

```python
y_pred_prob = classifier.predict_proba(X_test)[:, 1]
pd.DataFrame({'Churn': y_pred_prob}).to_csv(f'submission_logreg.csv', index_label='Id')
```

**Результат:**

ROC-AUC вашего решения равен 0.8458867918430664. Это дает вам 5.0 баллов.

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

(Дополнительно) Есть некоторые сложности с тем, чтобы использовать CatBoostClassifier вместе с GridSearchCV, поэтому мы не просим использовать кроссвалидацию. Но можете попробовать)

### 1. Заведем функцию для построения ROC-кривой

```python
def plot_roc_curve(y_train, y_train_pred,
                   y_valid, y_valid_pred,
                   model=None):

  train_score = roc_auc_score(y_train, y_train_pred)
  valid_score = roc_auc_score(y_valid, y_valid_pred)

  train_fpr, train_tpr, _ = roc_curve(y_train, y_train_pred)
  valid_fpr, valid_tpr, _ = roc_curve(y_valid, y_valid_pred)

  plt.figure(figsize=(12, 8))
  sns.lineplot(x=train_fpr, y=train_tpr, label=f'Train score: {train_score:2f}',)
  sns.lineplot(x=valid_fpr, y=valid_tpr, label=f'Valid score: {valid_score:2f}')
  sns.lineplot(x=np.linspace(0, 1, 100), y=np.linspace(0, 1, 100))

  plt.title(f'ROC curve for {model}')
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')

  plt.show()
```

### 2. Загрузим данные и применим градиентный бустинг

```python
X_df_train, y_df_train, X_df_test = load_data(FEATURE_CONFIG)
```

#### 2.1 Разделим данные на обучающую и валидационную части

```python
X_train, X_valid, y_train, y_valid = train_test_split(
    X_df_train,
    y_df_train,
    test_size=0.25,
    stratify=y_df_train,
    random_state=42
)
```

#### 2.2 Обучение и применение `CatBoostClassifier`

```python
train_data = Pool(X_train, y_train, cat_features=FEATURE_CONFIG['categorical'])
valid_data = Pool(X_valid, y_valid, cat_features=FEATURE_CONFIG['categorical'])
```

```python
cat_classifier = CatBoostClassifier(
    random_state=42,
    verbose=0,
    eval_metric='AUC'
)
```

```python
cat_classifier.fit(train_data, verbose=False)
```

```python
plot_roc_curve(y_train, cat_classifier.predict_proba(X_train)[:, 1],
               y_valid, cat_classifier.predict_proba(X_valid)[:, 1],
               model='CatBoostClassifier')
```

### 3 Предсказания градиентного бустинга
